```python
import earthaccess
import xarray as xr
```

```python
earthaccess.login()
```

### Test on MUR collection

```python
%%time
mapper_mur = earthaccess.get_virtual_reference(short_name = "MUR-JPL-L4-GLOB-v4.1")
```

```python
%%time
ds_mur = xr.open_dataset(mapper_mur, engine="zarr", chunks={}, backend_kwargs={"consolidated":False})
ds_mur
```

### Test on OSTIA collection

```python
%%time
mapper_ostia = earthaccess.get_virtual_reference(short_name = "OSTIA-UKMO-L4-GLOB-REP-v2.0", format="json")
```

```python
%%time
ds_ostia = xr.open_dataset(mapper_ostia, engine="zarr", chunks={}, backend_kwargs={"consolidated":False})
ds_ostia
```
